#### a) Obliczanie wartości wybranych opisowych charakterystyk rozkładu

Korzystając z danych zawartych w tablicy `recession_table_1` oblicz dla każdego roku następujące opisowe charakterystyki rozkładu:
- wartość średnia (ang. mean),
- odchylenie standardowe (ang. standart deviation),
- kwartyle (ang. quartiles),
- moda (ang. mode).

Zestaw wyniki w tablicy `descriptive_parameters_2`.

Indeks:
 - zawiera następujące nazwy wierszy: mean, std, Q1, Q2, Q3, mode.
 
Kolumny:
 - lata (osobna kolumna dla każdego roku opatrzona nagłówkiem 2006, 2007 itd. sformatowanym jako str)
 
Uwaga! W celu otrzymania maksymalnej liczby punktów otrzymane wyniki powinny być w całości wynikiem pracy napisanego kodu, bez manualnego określania, w którym przedziale znajdują się określone statystyki itp.

```python
# przygotowanie szkieletu DataFrame
years = recession_table_1.columns
descriptive_parameters_2 = pd.DataFrame(columns=years)

# zmienne pomocnicze
n = recession_table_1.sum()[0]
middles = recession_table_1.index.mid
beginnings = recession_table_1.index.left
q1_n, q2_n, q3_n = int(0.25 * n), int(0.5 * n), int(0.75 * n)
iq = 5

# średnia
descriptive_parameters_2.loc['mean']=(recession_table_1.multiply(middles, axis=0)).sum()/n

for year in recession_table_1.columns:
    year_col = recession_table_1[year]
    
    # odchylenie standardowe
    descriptive_parameters_2.loc['std', year] = np.sqrt((((middles - descriptive_parameters_2.loc['mean', year])**2)*year_col).sum()/n)
    
    # pierwszy kwantyl
    arg_q1 = np.argmax(np.where(q1_n<=year_col.cumsum(), 1, 0))
    xq = beginnings[arg_q1]
    descriptive_parameters_2.loc['Q1', year] = xq + (n/4 - np.sum(year_col.iloc[:arg_q1]))/year_col.iloc[arg_q1]*iq
    
    # drugi kwantyl (mediana)
    arg_q2 = np.argmax(np.where(q2_n<=year_col.cumsum(), 1, 0))
    xq = beginnings[arg_q2]
    descriptive_parameters_2.loc['Q2', year] = xq + (n/2 - np.sum(year_col.iloc[:arg_q2]))/year_col.iloc[arg_q2]*iq
    
    # trzeci kwantyl
    arg_q3 = np.argmax(np.where(q3_n<=year_col.cumsum(), 1, 0))
    xq = beginnings[arg_q3]
    descriptive_parameters_2.loc['Q3', year] = xq + (3*n/4 - np.sum(year_col.iloc[:arg_q3]))/year_col.iloc[arg_q3]*iq
    
    # moda
    arg_d = np.argmax(recession_table_1[year])
    xd = beginnings[arg_d]
    nd = recession_table_1[year].iloc[arg_d]
    nd_1 = recession_table_1[year].iloc[arg_d-1]
    nd_plus = recession_table_1[year].iloc[arg_d+1]
    descriptive_parameters_2.loc['mode', year] = xd + ((nd-nd_1)/((nd-nd_1)+(nd-nd_plus)))*iq
    
    
descriptive_parameters_2
```
