```python
# Całość danych
# YOUR CODE HERE

# stworzenie kopii datasetu, aby go nie popsuć
rd1 = recession_data_1.copy()

# stworzenie kolumn (interwałów)
cols = pd.interval_range(start=-20, end=35, freq=5)

# stworzenie szkieletu DataFrame
ix =  pd.Index(['2006', '2007', '2008', '2009', '2010', '2011'])
recession_table_1 = pd.DataFrame(index=ix, columns=cols)

# obliczenie częstości dla każdego interwału, roku i regionu
for ind in ix:
    recession_table_1.loc[ind] = recession_data_1[ind].value_counts(normalize=True, bins=cols)

# wyświetlanie
recession_table_1.style.format(precision=2)

# raise NotImplementedError()
```

```python
# Podział na rejony
# YOUR CODE HERE

# stworzenie szkieletu DataFrame
ix =  pd.MultiIndex.from_product([recession_data_1['Region Name'].unique(), ['2006', '2007', '2008', '2009', '2010', '2011']])
recession_table_2 = pd.DataFrame(index=ix, columns=cols)

# obliczenie częstości dla każdego interwału, roku i regionu
for ind in ix:
    recession_table_2.loc[ind] = recession_data_1[recession_data_1['Region Name'] == ind[0]][ind[1]].value_counts(normalize=True, bins=cols)

# wyświetlanie
recession_table_2.style.format(precision=2)

# raise NotImplementedError()
```

```python
# Podział pod względem stopnia zamożności
# YOUR CODE HERE

# stworzenie szkieletu DataFrame
ix =  pd.MultiIndex.from_product([recession_data_1['Income Group'].unique(), ['2006', '2007', '2008', '2009', '2010', '2011']])
recession_table_3 = pd.DataFrame(index=ix, columns=cols)

# obliczenie częstości dla każdego interwału, roku i grupy dochodowej
for ind in ix:
    recession_table_3.loc[ind] = recession_data_1[recession_data_1['Income Group'] == ind[0]][ind[1]].value_counts(normalize=True, bins=cols)

# wyświetlanie
recession_table_3.style.format(precision=2)

# raise NotImplementedError()
```

## Wykres boxplot
#### Zalety
1. Zawiera najważniejsze informacje o danych: medianę, kwartyle, wartości odstające, wartości min-max oraz IQR
2. Jest najbardziej przejrzysty
3. Najszybszy do wykonania, co w przypadku wielu danych ma duże znaczenie
4. Widać wartości odstające
#### Wady
1. Nie widać ogólnego rozkładu danych
2. Skośność danych nie jest pokazana wprost, trzeba jej się domyślić
3. Nie widzimy średniej danych
#### Problemy
1. Jeśli dane będą mocno skośne taki wykres może nie wyglądać zbyt dobrze - będzie zbyt wydłużony
2. Pokazując je osobom spoza środowiska mogą nastręczać problemy w interpretacji i wprowadzać w błąd
## Wykres violinplot
#### Zalety
1. Pokazuje również rozkład danych
2. Atrakcyjniejszy wizualnie
3. Przejrzysty (chociaż w mojej opinii mniej niż boxplot)
#### Wady
1. Jest jeszcze bardziej skomplikowany od boxplotu
2. Wymaga większej ilości danych aby był użyteczny
3. Nie pokazuje wprost wartości odstających
#### Problemy
1. Posiadając niewiele wartości odstających zostanie to źle zobrazowane na takim wykresie
## Wykres swarmplot
#### Zalety
1. Pokazuje dokładny rozkład danych
2. Dobrze pokazuje wartości odstające
#### Wady
1. Nie pokazuje takich informacji jak mediana, średnia, IQR i innych
2. Długo się liczy
#### Problemy
1. Dla dużej ilości danych będzie się długo liczył, a wyniki mogą nie być przejrzyste
## Wykres stripplot
#### Zalety
1. Pokazuje dokładne położenie danych
2. Stosunkowo prosty, prawdopodobnie najbardziej intuicyjny z wszystkich wykresów do wizualizacji rozkładu danych
#### Wady
1. Nie pokazuje mediany średniej, IQR, wartości min i max itd.
2. Nie posiada "zgrubień" tam gdzie występuje wiele danych więc może być mylący
3. Długo się liczy
#### Problemy
1. Dla dużej ilości danych dane będą się na siebie nakładać i niewiele będzie można z niego odczytać
2. Bez innych wykresów nie dowiemy się zbyt wiele o danych
