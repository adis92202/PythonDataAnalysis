#### a) Wizualizacja

Zwizualizuj funkcje gęstości prawdopodobieństwa następujących rozkładów teoretycznych:

#### b) Charakterystyka rozkładów
Na podstawie przygotowanych wykresów scharakteryzuje opisane rozkłady teoretyczne i odpowiedz na pytanie w jaki sposób zmiana wartości parametrów rozkładów wpływa na położenie i kształt ich funkcji gęstości prawdopodobieństwa.

#### Rozkład normalny
Rozkład normalny (Gaussa), a dokładniej jego funkcja gęstości prawdopodobieństwa jest dzwonowatego kształtu, symetryczny względem środka (średniej). Zmiana średniej zmienia punkt w którym występuje maksimum funkcji, przesuwa go wizualnie na wykresie. Odchylenie standardowe determinuje "szerokość" wykresu - im większe tym wykres będzie bardziej rozległy, rozlany, będzie wolniej dążył do zera. Mniejsze wartości odchylenia standardowego powodują koncentrację wykresu wokół średniej.

#### Rozkład t-Studenta
W przeciwieństwie do rozkładu normalnego jego środek (średnia, punkt maksymalny) zawsze znajduje się w punkcie 0. Również jest symetryczny a jego kształt w przybliżeniu dzwonowaty. Im większa ilość stopni swobody tym bardziej zbliża się on wyglądem do funkcji gęstości prawdopodobieństwa dla N(0,1). 

#### Rozkład chi-kwadrat
Rozkład ten określony jest tylko dla x>0. Jego kształt jest w przybliżeniu podobny do rozkładu wykładniczego. Posiada jeden parametr oznaczający ilość stopni swobody. Im większa wartość tego parametru tym bardziej wykres jest spłaszczony, a wartość maksymalna przesuwa się w prawo.

#### Rozkład F Snedecora
Ten rozkład również określony jest tylko dla x>0. Jego kształt przypomina rozkład chi-kwadrat, jednak posiada wyraźniejsze peaki (wartości maksymalne). Posiada dwa parametry, określane jako *d1,d2 stopni swobody*. Wraz ze zwiększaniem się pierwszego parametru *peak* na wykresie przesuwa się w prawo i spłaszcza się, choć nie tak wyraźnie jak w rozkładzie chi-kwadrat. Wraz ze zwiększaniem się drugiego parametru zwiększa się ogólny poziom wartości na wykresie.


### Zadanie 2: Praca z rozkładami teoretycznymi [10 pkt]

#### a) Wyliczanie prawdopodobieństw na podstawie wartości

Oblicz i zapisz do zmiennych o podanych nazwach następujące wartości prawdopodobieństwa:
- `p1` - $P(X < -1.5)$, gdzie $X$ to zmienna losowa ciągła posiadająca rozkład t-Studenta o 10 stopniach swobody - $t(10)$,
- `p2` - $P(X > 2)$, gdzie $X$ to zmienna losowa ciągła posiadająca rozkład F Snedecora o 5 i 50 stopniach swobody - $F(5, 50)$,
- `p3` - $P(3 < X < 10)$, gdzie $X$ to zmienna losowa ciągła posiadająca rozkład chi-kwadrat o 4 stopniach swobody - $\chi^2(4)$,
- `p4` - $P(X < -2.75 | X > 1.5)$, gdzie $X$ to zmienna losowa ciągła posiadająca rozkład normalny o wartości oczewiwanej równej -0.5 i odchyleniu standardowym równym 1.5 - $N(-0.5, 1.5)$,

Każde z rozwiązań zilustruj wykresem zawierającym krzywą gęstości prawdopodobieństwa rozkładu z zaznaczonym kolorem fragmentem pola pod krzywą odpowiadający wyznaczonej wartości prawdopodobieństwa. Przygotuj figurę zgodnie z wytycznymi z nagłówka notatnika. Dodatkowo:
- ustaw zakres osi x od wartości zmiennej $X$, dla której dystrybuanta analizowanego rozkładu przyjmuje wartość 0.0001 do wartości zmiennej $X$, dla której dystrybuanta analizowanego rozkładu przyjmuje wartość 0.9999,
- w tytule wykresu zawrzyj uzupełniony wzór na prawdopodobieństwo uzupełniony o wyliczoną wartość prawdopodobieństwa podaną do 4 miejsca po przecinku (np. P(X < -1.5) = x.xxxx, gdzie x.xxxx to obliczona wartość).

```python
# p1
# YOUR CODE HERE

# Obliczenie p1
p1 = sp.stats.t.cdf(-1.5, 10)
print(p1)

# Wyznaczenie danych do wykresu
xmin, xmax = sp.stats.t.ppf(0.0001,10), sp.stats.t.ppf(0.9999, 10)
x = np.linspace(xmin, xmax, 1000)
y = sp.stats.t.pdf(x, 10)

# Wykres
fig, ax = plt.subplots(1,1,figsize=(10,8))
sns.lineplot(x=x, y=y, ax=ax)
ax.set(xlim=[xmin,xmax], ylim=[-0.001,1], title=f'P(X < -1.5) = {p1:.4f}', xlabel='X', ylabel='pdf')
ax.fill_between(x, y1=0, y2=y, where=x<-1.5, color="grey")
plt.show()

# raise NotImplementedError()
```

```python
# p2
# YOUR CODE HERE

# Obliczenie p2
p2 = 1-sp.stats.f.cdf(2, 5, 50)
print(p2)

# Wyznaczenie danych do wykresu
xmin, xmax = sp.stats.f.ppf(0.0001, 5, 50), sp.stats.f.ppf(0.9999, 5, 50)
x = np.linspace(xmin, xmax, 1000)
y = sp.stats.f.pdf(x, 5, 50)

# Wykres
fig, ax = plt.subplots(1,1,figsize=(10,8))
sns.lineplot(x=x, y=y, ax=ax)
ax.set(xlim=[xmin,xmax], ylim=[-0.001,1], title=f'P(X > 2) = {p2:.4f}', xlabel='X', ylabel='pdf')
ax.fill_between(x, y1=0, y2=y, where=x>2, color="grey")
plt.show()

# raise NotImplementedError()
```

```python
# p3
# YOUR CODE HERE

# Obliczenie p3
p3_1 = sp.stats.chi2.cdf(3, 4)
p3_2 = sp.stats.chi2.cdf(10, 4)
p3 = p3_2-p3_1
print(p3)

# Wyznaczenie danych do wykresu
xmin, xmax = sp.stats.chi2.ppf(0.0001, 4), sp.stats.chi2.ppf(0.9999, 4)
x = np.linspace(xmin, xmax, 1000)
y = sp.stats.chi2.pdf(x, 4)

# Wykres
fig, ax = plt.subplots(1,1,figsize=(10,8))
sns.lineplot(x=x, y=y, ax=ax)
ax.set(xlim=[xmin,xmax], ylim=[-0.001,1], title=f'P(3 < X < 10) = {p3:.4f}', xlabel='X', ylabel='pdf')
ax.fill_between(x, y1=0, y2=y, where=((x>3) & (x<10)), color="grey")
plt.show()

# raise NotImplementedError()
```

```python
# p4
# YOUR CODE HERE

# Obliczenie p4
p4_1 = sp.stats.norm.cdf(-2.75, -0.5, 1.5)
p4_2 = 1-sp.stats.norm.cdf(1.5, -0.5, 1.5)
p4 = p4_2+p4_1
print(p4)

# Wyznaczenie danych do wykresu
xmin, xmax = sp.stats.norm.ppf(0.0001, -0.5, 1.5), sp.stats.norm.ppf(0.9999, -0.5, 1.5)
x = np.linspace(xmin, xmax, 1000)
y = sp.stats.norm.pdf(x, -0.5, 1.5)

# Wykres
fig, ax = plt.subplots(1,1,figsize=(10,8))
sns.lineplot(x=x, y=y, ax=ax)
ax.set(xlim=[xmin, xmax], ylim=[-0.001,1], title=f'P(X < -2.75|X > 1.5) = {p4:.4f}', xlabel='X', ylabel='pdf')
ax.fill_between(x, y1=0, y2=y, where=((x>1.5) | (x<-2.75)), color="grey")
plt.show()

# raise NotImplementedError()
```

#### b) Wyliczanie wartości na podstawie prawdopodobieństw

Rozwiąż i zapisz do zmiennych o podanych nazwach rozwiązania następujących równań:
- `s1` - $P(X < s) = 0.25$, gdzie X to zmienna losowa ciągła posiadająca rozkład chi-kwadrat o 5 stopniach swobody - $\chi^2(5)$,
- `s2` - $P(X > s) = 0.15$, gdzie X to zmienna losowa ciągła posiadająca rozkład F Snedecora o 5 i 20 stopniach swobody - $F(5, 20)$,
- `s3` - $P(-1.25 < X < s) = 0.35$, gdzie x to zmienna losowa ciągła posiadająca rozkład normalny o wartości oczewiwanej równej -0.4 i odchyleniu standardowym równym 0.75 - $N(-0.4, 0.75)$,

Każde z rozwiązań zilustruj wykresem zawierającym krzywą gęstości prawdopodobieństwa rozkładu z zaznaczonym kolorem fragmentem pola pod krzywą odpowiadający wartości prawdopodobieństwa określonej przez rozwiązywane równanie. Przygotuj figurę zgodnie z wytycznymi z nagłówka notatnika. Dodatkowo:
- ustaw zakres osi x od wartości zmiennej $X$, dla której dystrybuanta analizowanego rozkładu przyjmuje wartość 0.0001 do wartości zmiennej $X$, dla której dystrybuanta analizowanego rozkładu przyjmuje wartość 0.9999,
- oznacz krawędź obszaru, na której znajduje się wyznaczona wartość $s$ linią w wyróżniającym się kolorze,
- w tytule wykresu zawrzyj uzupełniony wzór na prawdopodobieństwo uzupełniony o wyliczoną wartość zmiennej $X$ podaną do 4 miejsca po przecinku (np. P(X < s.ssss) = 0.25, gdzie s.ssss to obliczona wartość).

```python
# s1
# YOUR CODE HERE

# Obliczenie s1
s1 = sp.stats.chi2.ppf(0.25, 5)
print(s1)

# Wyznaczenie danych do wykresu
xmin, xmax = sp.stats.chi2.ppf(0.0001, 5), sp.stats.chi2.ppf(0.9999, 5)
x = np.linspace(xmin, xmax, 1000)
y = sp.stats.chi2.pdf(x, 5)

# Wykres
fig, ax = plt.subplots(1,1,figsize=(10,8))
sns.lineplot(x=x, y=y, ax=ax)
ax.set(xlim=[xmin,xmax], ylim=[-0.001,1], title=f'P(X < {s1:.4f}) = 0.25', xlabel='X', ylabel='pdf')
ax.fill_between(x, y1=0, y2=y, where=x<s1, color="grey")
ax.vlines(s1, 0, sp.stats.chi2.pdf(s1, 5), colors='red')
plt.show()

# raise NotImplementedError()
```

```python
# s2
# YOUR CODE HERE

# Obliczenie s2
s2 = sp.stats.f.ppf(1-0.15, 5, 20)
print(s2)

# Wyznaczenie danych do wykresu
xmin, xmax = sp.stats.f.ppf(0.0001, 5, 20), sp.stats.f.ppf(0.9999, 5, 20)
x = np.linspace(xmin, xmax, 1000)
y = sp.stats.f.pdf(x, 5, 20)

# Wykres
fig, ax = plt.subplots(1,1,figsize=(10,8))
sns.lineplot(x=x, y=y, ax=ax)
ax.set(xlim=[xmin,xmax], ylim=[-0.001,1], title=f'P(X > {s2:.4f}) = 0.15', xlabel='X', ylabel='pdf')
ax.fill_between(x, y1=0, y2=y, where=x>s2, color="grey")
ax.vlines(s2, 0, sp.stats.f.pdf(s2, 5, 20), colors='red')
plt.show()

# raise NotImplementedError()
```

```python
# s3
# YOUR CODE HERE

# Obliczenie s3
p_temp = sp.stats.norm.cdf(-1.25, -0.4, 0.75)
s3 = sp.stats.norm.ppf(p_temp+0.35, -0.4, 0.75)
print(s3)

# Wyznaczenie danych do wykresu
xmin, xmax = sp.stats.norm.ppf(0.0001, -0.4, 0.75), sp.stats.norm.ppf(0.9999, -0.4, 0.75)
x = np.linspace(xmin, xmax, 1000)
y = sp.stats.norm.pdf(x, -0.4, 0.75)

# Wykres
fig, ax = plt.subplots(1,1,figsize=(10,8))
sns.lineplot(x=x, y=y, ax=ax)
ax.set(xlim=[xmin,xmax], ylim=[-0.001,1], title=f'P(-1.25 < X < {s3:.4f}) = 0.35', xlabel='X', ylabel='pdf')
ax.fill_between(x, y1=0, y2=y, where=((x>-1.25) & (x<s3)), color="grey")
ax.vlines(s3, 0, sp.stats.norm.pdf(s3, -0.4, 0.75), colors='red')
ax.vlines(-1.25, 0, sp.stats.norm.pdf(-1.25, -0.4, 0.75), colors='red')
plt.show()

# raise NotImplementedError()
```
